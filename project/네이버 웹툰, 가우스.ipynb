1. 네이버 웹툰 스크래핑

```python
#항상 복붙한다고 생각하기
import requests
from bs4 import BeautifulSoup
url = "https://comic.naver.com/webtoon/weekday.nhn"
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36","Accept-Language": "ko-KR,ko;q=0.8,en-US;q=0.5,en;q=0.3"}
res = requests.get(url, headers= headers)
res.raise_for_status()

soup = BeautifulSoup(res.text, "lxml")
```

#1. 형제들을 찾고 싶을 때, 연애혁명 다음 rank를 프린트하고 싶을 때 

```python
rank1 = soup.find("li" ,attrs = {"class" : "rank01"})
print(rank1.a.get_text())
print(rank1.next_sibling)

#만약 출력이 안되면 next_sibling을 한번 더 쳐보기 -> 애들 사이에 줄바꿈이 있으면 원하는 내용이 출력 안될 수도
#다음 형제는 next_sibling인데, 기존 형제는 previous_sibling임
#부모는 rank1.paraent하면 됨
#next_sibling을 하는데 조건을 넣어서 찾고싶으면 괄호 열고 조건 적어주기
#rank1.next_sibling("li") -> 이러면 줄바꿈 고려 안해도 됨!
#형제들이 여럿일 경우 next_siblings << s붙이면 됨
```

#3. 만화 회차 제목 가져오기

```python
#좋아하는 웹툰 암거나 찍어보겠음 최근에 재밌게 본 '시월드가 내게 집착한다'를 들어감
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://comic.naver.com/webtoon/list?titleId=785251&weekday=sun&page=1" #시월드 url
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36","Accept-Language": "ko-KR,ko;q=0.8,en-US;q=0.5,en;q=0.3"}
res = requests.get(url, headers= headers)
res.raise_for_status()

soup = BeautifulSoup(res.text, "lxml")
```

```python
#페이지에서 f12를 열고 보면 tbody아래에 tr들로 각 화를 관리하는 것을 알 수 있음
items = soup.find_all("tr")
print(items)
#그러면 회차도 나오는데 평점 등 이상한거도 나옴 근데 td class="title" 아래에 회차가 있는 걸 알 수 있음
```

```python
items = soup.find_all("td",attrs={"class":"title"})
print(items)
```

```python
for x in items :
    print(x.get_text().strip())
#strip중요, 근데 일부회차만 나옴 -> page=1이라서 그럼
```

```python
#최종, url page때문에 for문 돌리고 나머지 똑같음
for x in range(1,7) :
    url = "https://comic.naver.com/webtoon/list?titleId=785251&weekday=sun&page={}".format(x) #시월드 url
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36","Accept-Language": "ko-KR,ko;q=0.8,en-US;q=0.5,en;q=0.3"}
    res = requests.get(url, headers= headers)
    res.raise_for_status()

    soup = BeautifulSoup(res.text, "lxml")
    items = soup.find_all("td",attrs={"class":"title"})

    for x in items :
        print(x.get_text().strip())
```

#5. <평점 가져오기>

```python
import requests
from bs4 import BeautifulSoup

for i in range(7) :
    url = "https://comic.naver.com/webtoon/list?titleId=785251&weekday=sun&page={}".format(i) #시월드 url
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36","Accept-Language": "ko-KR,ko;q=0.8,en-US;q=0.5,en;q=0.3"}
    res = requests.get(url, headers= headers)
    res.raise_for_status()

    soup = BeautifulSoup(res.text, "lxml")

    rating = soup.find_all("div", attrs = {"class":"rating_type"})

    for ratings in rating: 
        print(ratings.get_text().strip())
```
